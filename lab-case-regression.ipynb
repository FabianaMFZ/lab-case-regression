### T-test to compare category 0 and category 1 for target_b

```python
"""Performing the T-Test: 
For each feature column, an independent two-sample t-test (ttest_ind) is performed to compare the means of the two groups.
Storing and Filtering Results: 
The results (t-statistic and p-values) are stored, converted to a DataFrame, and then filtered to show only those features 
with significant differences (p-value <= 0.05)."""

target_0 = df_target_0.drop(['target_d'], axis=1)
target_1 = df_target_1.drop(['target_d'], axis=1)


import scipy.stats as stats

results = []

# Iterate over each column to perform t-test
for column in target_0.columns:
    # Perform the t-test
    t_stat, p_value = stats.ttest_ind(target_0[column], target_1[column], equal_var=False)
    
    # Append results to the list
    results.append((column, t_stat, p_value))

# Convert results to a DataFrame
results_df = pd.DataFrame(results, columns=['Column', 'T-Statistic', 'P-Value'])
results_df
```

```python
# Filter the results to get only significant columns (p-value <= 0.05)
significant_results = results_df[results_df['P-Value'] <= 0.05]
significant_results
```

```python
significant_results['Column'].unique()
```

#### Why It Matters for Feature Selection

By using the p-value from the t-test, you can pre-select features that show significant differences between your target groups. This helps you focus on features that are more likely to be relevant for predicting the outcome, improving the performance and interpretability of your logistic regression model.

```python
y = df_target_1['target_d']
X = df_target_1.drop(['target_d'], axis = 1)
```

#### Decision Tree

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size = 0.2)

from sklearn.tree import DecisionTreeRegressor

max_depth = range(1,30)
test = []
train = []

for depth in max_depth:
    model = DecisionTreeRegressor(max_depth= depth)
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    mse = mean_squared_error (y_test, pred)
    test.append(np.sqrt(mse))
```

```python
plt.plot( test, label="test error")
plt.ylabel("RMSE")
plt.xlabel("n_depth")
plt.legend()
```

```python
tree_model = DecisionTreeRegressor(max_depth = 2)
tree_model.fit(X_train, y_train)
```

```python
tree_pred = tree_model.predict(X_test)

from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

print("R2 value is = ",round(r2_score(y_test, tree_pred),2))
mse = mean_squared_error (y_test, tree_pred)
mae = mean_absolute_error (y_test, tree_pred)
print("The mse of the model is = ",mse)
print("The root mse of the model is = ",np.sqrt(mse))
print("The mean absolute error of the model is = ",mae)
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV 
from sklearn.model_selection import KFold

model = RandomForestRegressor()

param_grid = {
    'n_estimators': [20,50],
    'max_depth': [2,3],
    'min_samples_split': [2, 4],
    'min_samples_leaf' : [1, 2],
    'bootstrap': [True, False]
    }

# Use KFold for cross-validation
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=kfold, scoring='neg_mean_squared_error', verbose=2, n_jobs=-1)

# Fit the model
grid_search.fit(X_train, y_train)

# Best model
best_model = grid_search.best_estimator_

# Predict on test data
y_pred = best_model.predict(X_test)

# Evaluate the model
rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f'RMSE: {rmse}')
```
